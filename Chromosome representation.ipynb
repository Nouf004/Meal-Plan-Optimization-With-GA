```python
#بسم الله الرحمن الرحيم وبه نستعين
# We import dataclass to represent 1 gene 
from dataclasses import dataclass 
# To read the dataset
import csv
import random
```

```python
# A gene contains 2 attributes : the food item's name and the number of servings of this item
@dataclass
class Gene:
    food_Item_Name: str
    serving: int
```

```python
# We make a list representing all the meals we have in 1 chromosome 
Meals =["Breakfast", "Snack1", "Lunch", "Snack2", "Dinner"]

# We make a list representing all the food groups we have in 1 meal
Food_Groups =["Vegetables", "Fruits", "Grains", "Protein", "Dairy", "Fats"]
```

```python
# Here each one of the 6 categories will have their own array that will be taken from the dataset
#Food_Dataset = {
   # group: [] for group in Food_Groups
#}

# this is just a place holder
Food_Dataset = {
    "Vegetables": ["Tomato", "Carrot", "Cucumber"],
    "Fruits": ["Apple", "Banana", "Orange"],
    "Grains": ["Rice", "Bread", "Pasta"],
    "Protein": ["Chicken", "Fish", "Eggs"],
    "Dairy": ["Milk", "Yogurt", "Cheese"],
    "Fats": ["Olive Oil", "Butter"]
}

```

```python
# Here we will read from the csv 
```

```python
# We make the chromosome structure in which we have 5 meals that contain 6 genes each 
chromosome = [[None for _ in range(6)] for _ in range(5)]

def initialize_chromosome():
    for i in range(len(Meals)):
        for j in range(len(Food_Groups)):
            group = Food_Groups[j]
            food_name = random.choice(Food_Dataset[group])
            serving = random.uniform(50, 200)

            chromosome[i][j] = Gene(food_name, serving)

    return chromosome

# to print the chromosome for checking 
def print_chromosome(chromosome):
    for i in range(len(Meals)):
        print(f"\n{Meals[i]}")
        for j in range(len(Food_Groups)):
            gene=chromosome[i][j]
            print(f"  {Food_Groups[j]} -> {gene.food_Item_Name}, {gene.serving:.1f}")
```

```python
# Initialize the chromosome 
chromosome=initialize_chromosome()

print_chromosome(chromosome)
```
